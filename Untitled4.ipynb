```python
# khởi tạo
first = [11.25, 18.0, 20.0]
second = [10.75, 9.50]

# ghép first và second vào danh sách full
full = first + second

# sắp xếp lại full theo thứ tự giảm dần, lưu vào biến full_sorted
full_sorted = sorted(full, reverse=True)

# in ra full_sorted
print(full_sorted)
```
